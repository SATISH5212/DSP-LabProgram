```python
#lineplot
import numpy as np
import matplotlib.pyplot as plt
x=np.array([0,1,2,3,4,5])
y=x*2


plt.title("line plot graph with lables,title,markes ,lineshapes,colors,grids")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y,marker="*",mec="red")
plt.grid()
plt.show()
```

```python
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(1,2,2)
plt.plot(x,y,"-.",marker="*")

x = np.array([0,1,2,1])
y = np.array([10,20,30,40])

plt.subplot(1,2,2)
plt.plot(x,y)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(1,2,1)
plt.plot(x,y,"-.",marker="*")

x = np.array([0,1,2,1])
y = np.array([10,20,30,40])

plt.subplot(1,2,2 )
plt.plot(x,y,mec="green")

plt.show()
```

```python
#piechart
x=np.array([30,20,10,5,35])
plt.pie(x)
plt.show()
```

```python
x=np.array([30,20,10,5,35])
lab=["A","B","C","D","E"]
plt.pie(x,labels=lab)
plt.show()
```

```python
x=np.array([30,20,10,5,35])
lab=["A","B","C","D","E"]
plt.pie(x,labels=lab)
plt.legend()
plt.show()
```

```python
x=np.array([30,20,10,5,35])
lab=["A","B","C","D","E"]
col=["red","blue","green","yellow","white"]
plt.pie(x,labels=lab,colors=col)
plt.legend(title="sum")
plt.show()
```
