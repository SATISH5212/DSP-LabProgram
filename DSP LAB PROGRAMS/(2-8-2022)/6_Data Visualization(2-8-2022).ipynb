```python
import pandas as pd
df=pd.read_csv("sales.csv")
import matplotlib.pyplot as plt
print(df)

user=df["month_number"].tolist()
follower=df["total_profit"].tolist()
plt.plot(user,follower,"--")
plt.xlabel("Month number")
plt.ylabel("total profit")
plt.title("company profit per month")
plt.show()
```

```python
plt.plot(user,follower,color="r",label="profit data of last year",marker="o",linewidth=3,linestyle="--",mec="black")
plt.xlabel("Month number")
plt.ylabel("total profit")
plt.title("company profit per month")
plt.legend(loc="lower right")
plt.show()
```
